### Q1) Create a GUI based application to add two numbers and get the result using Gooeypie

```python
# importing all the necessary libraries/modules
import gooeypie as gp
```

```python
def addNumbers(event):
    try:
        num1 = float(input1.text)
        num2 = float(input2.text)
        result = num1 + num2
        
        submit_label.text = f"Result: {result:.2f}"  # Display result with two decimal places
        
    except ValueError:
        submit_label.text = "Invalid input. Please enter valid numbers."
```

```python
# Initiating the application
app = gp.GooeyPieApp("Addition")

# Setting the width of the window
app.width = 275
app.height = 275

# Taking user input
input1_label = gp.Label(app, "Enter the first number:")
input1 = gp.Input(app)

input2_label = gp.Label(app, "Enter the second Number:")
input2 = gp.Input(app)

# Submit button
submit = gp.Button(app, 'ADD', addNumbers)
submit_label = gp.Label(app, "")

# Button positioning
app.set_grid(6, 3)

app.add(input1_label, 1, 1)
app.add(input1, 1, 3)

app.add(input2_label, 2, 1)
app.add(input2, 2, 3)

app.add(submit, 3, 2)
app.add(submit_label, 5, 2)

# running the application
app.run()
```

### Q2) Create a simple feedback form using Gooeypie

```python
# Initiating the application
app = gp.GooeyPieApp('Feedback Form')

# Setting the width of the window
app.width = 275
app.height = 275

# name widget
name_lbl = gp.Label(app, 'Name:')
name = gp.Input(app)

# feedback text box
feedback_lbl = gp.Label(app, 'Please provide your feedback:')
feedback = gp.Input(app) 

# Create a label for displaying messages
message_label = gp.Label(app, '')

# User-Defined function for button click
def submit_feedback(event):
    message_label.text = "Thank you for your feedback"

# submit button
submit = gp.Button(app, 'Submit', submit_feedback)

# Button positioning
app.set_grid(6, 1)  
app.add(name_lbl, 1, 1)
app.add(name, 2, 1, fill=True)
app.add(feedback_lbl, 3, 1)
app.add(feedback, 4, 1, fill=True)
app.add(submit, 5, 1)
app.add(message_label, 6, 1)  

# running the application
app.run()
```
