```python
# importing all the necessary libraries/modules
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

**Q1) Uploading a group photo**

```python
# uploading the image
image = cv2.imread('D:/Z/Downloads/group_photo.jpg')

# RGB Image
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
fig1, ax = plt.subplots(1, 2, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image)
ax[0].set_title("Original Image")

ax[1].imshow(image_RGB)
ax[1].set_title("RGB Image")

plt.show()
```

**Q2) Cropping an object from the image**

```python
# Define the coordinates of the region of interest (ROI)
x = 0  # starting x-coordinate
y = 80  # starting y-coordinate
width = 180  # width of the ROI
height = 200  # height of the ROI

# Extract the ROI from the image
cropped_image = image_RGB[y:y+height, x:x+width]
```

```python
plt.title("Cropped Image")
plt.xlabel("X-axis Scaling")
plt.ylabel("Y-axis Scaling")
plt.imshow(cropped_image)
plt.show()
```

**Q3) Show the different stages as images**

**Masking**

```python
# Setting up the lower and upper values of the mask
upper = np.array([255, 255, 255])
lower1 = np.array([223, 177, 150])
lower2 = np.array([205, 173, 150])
lower3 = np.array([215, 174, 56])
lower4 = np.array([182, 124, 94])
#lower5 = np.array([])

# Creating a mask of an image
mask1 = cv2.inRange(cropped_image, lower1, upper)
mask2 = cv2.inRange(cropped_image, lower2, upper)
mask3 = cv2.inRange(cropped_image, lower3, upper)
mask4 = cv2.inRange(cropped_image, lower4, upper)

# Creating a new image with layer over layer
masked_image1 = np.copy(cropped_image)
masked_image1[mask1 != 0] = [0, 255, 0] # [Red, Green, Blue]

# Creating a new image with layer over layer
masked_image2 = np.copy(cropped_image)
masked_image2[mask2 != 0] = [0, 255, 0] # [Red, Green, Blue]

# Creating a new image with layer over layer
masked_image3 = np.copy(cropped_image)
masked_image3[mask3 != 0] = [0, 255, 0] # [Red, Green, Blue]

# Creating a new image with layer over layer
masked_image4 = np.copy(cropped_image)
masked_image4[mask4 != 0] = [0, 255, 0] # [Red, Green, Blue]
```

```python
plt.subplot(421),plt.imshow(mask1, cmap = 'gray'), plt.axis('off')
plt.subplot(422),plt.imshow(masked_image1, cmap = 'gray'),plt.axis('off')

plt.subplot(423),plt.imshow(mask2, cmap = 'gray'), plt.axis('off')
plt.subplot(424),plt.imshow(masked_image2, cmap = 'gray'),plt.axis('off')

plt.subplot(425),plt.imshow(mask3, cmap = 'gray'), plt.axis('off')
plt.subplot(426),plt.imshow(masked_image3, cmap = 'gray'),plt.axis('off')

plt.subplot(427),plt.imshow(mask4, cmap = 'gray'), plt.axis('off')
plt.subplot(428),plt.imshow(masked_image4, cmap = 'gray'),plt.axis('off')

plt.show()
```

**Sharpening the image**

```python
# Defining the kernel 
kernel = np.array([[0, -1, 0],
                  [-1, 5, -1],
                  [0, -1, 0]])
```

```python
image_sharpened = cv2.filter2D(cropped_image, -1, kernel)
```

```python
plt.title("Sharpened Image")
plt.xlabel("X-axis Scaling")
plt.ylabel("Y-axis Scaling")
plt.imshow(image_sharpened)
plt.show()
```
