# `Cascade Classifier`

### Method 1: Detecting face via a given input

```python
face_cascade = cv2.CascadeClassifier('D:/Z/Downloads/haarcascade_frontalface_default.xml')
```

```python
# For an image
image = cv2.imread('D:/Z/Downloads/christianbale.png')
```

```python
# For an image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

### Method 2: Detecting face via live camera feed

```python
face_cascade = cv2.CascadeClassifier('D:/Z/Downloads/haarcascade_frontalface_default.xml')
```

```python
video_capture = cv2.VideoCapture(0)  # 0 for the default camera
```

```python
while True:
    ret, frame = video_capture.read()
    
    if not ret:
        break
    
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 255, 255), 2)
        
        # Labeling the face
        label = "Bomber"
        label_position = (x, y-10)  # Positioning the label just above the rectangle
        font_scale = 0.5
        font_color = (255, 255, 255)  # White color for the text
        font_thickness = 1
        font = cv2.FONT_HERSHEY_SIMPLEX

        cv2.putText(frame, label, label_position, font, font_scale, font_color, font_thickness)
    
    cv2.imshow('Face Recognition', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```

### Method 3: Detecting face via a given video

```python
import cv2
face_cascade = cv2.CascadeClassifier('D:/Z/Downloads/haarcascade_frontalface_default.xml')
```

```python
# Ask the user for input
source = input("Press 'c' for camera feed or 'v' for video file: ")

if source == 'c':
    video_capture = cv2.VideoCapture(0)  # 0 for the default camera
elif source == 'v':
    video_path = input("Enter the path to the video file: ")
    video_capture = cv2.VideoCapture(video_path)
else:
    print("Invalid option.")
    exit()

while True:
    ret, frame = video_capture.read()

    if not ret:
        break

    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)

        # Labeling the face
        label = "Bomber"
        label_position = (x, y-10)  # Positioning the label just above the rectangle
        font_scale = 0.5
        font_color = (255, 255, 255)  # White color for the text
        font_thickness = 1
        font = cv2.FONT_HERSHEY_SIMPLEX

        cv2.putText(frame, label, label_position, font, font_scale, font_color, font_thickness)

    cv2.imshow('Face Recognition', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```
