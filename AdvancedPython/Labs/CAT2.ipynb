**Q1) Write a Python program that reads an image and applies**
- Canny Edge Detection filter using OpenCV
- Sharpening image
- Edge Detection

```python
# installing the necessary libraries/modules
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# importing an image onto python
image = cv2.cvtColor(cv2.imread("C:/Users/Christ/Downloads/Images/image8.jpg"), cv2.COLOR_BGR2RGB)

# displaying the same image
plt.imshow(image)
plt.title("Image 8")
plt.xlabel("X-axis Scaling")
plt.ylabel("Y-axis Scaling")
plt.show()
```

`Sharpening the image`

```python
# defining the kernel 
kernel = np.array([
    [-1, -1, -1],
    [-1, 9, -1],
    [-1, -1, -1]
])

# sharpening the image via filter2D kernel
sharpened_image = cv2.filter2D(image, -1, kernel)

# sub-plotting the original and the sharpened image
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image'), plt.axis('off')
plt.subplot(122), plt.imshow(sharpened_image, cmap='gray'), plt.title('Sharpened Image'), plt.axis('off')
plt.show()
```

`Canny Edge Detection`

```python
# Detecting the edges through the canny edge detector
canny_image = cv2.Canny(image, 100, 200)

# sub-plotting the original and the sharpened image
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image'), plt.axis('off')
plt.subplot(122), plt.imshow(canny_image, cmap='gray'), plt.title('Sharpened Image'), plt.axis('off')
plt.show()
```

`Edge Detection`

```python
# sobel edge detection along the x-axis
sobel_x = cv2.Sobel(image, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=7)
# sobel edge detection along the y-axis
sobel_y = cv2.Sobel(image, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)
# sobel edge detection along both the axes
sobel_axes = cv2.Sobel(image, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3)
```

```python
# sub-plotting the original and the sharpened image
plt.subplot(221), plt.imshow(image, cmap='gray'), plt.title('Original Image'), plt.axis('off')
plt.subplot(222), plt.imshow(sobel_axes, cmap='gray'), plt.title('Along both axes'), plt.axis('off')
plt.subplot(223), plt.imshow(sobel_x, cmap='gray'), plt.title('Along X-axis'), plt.axis('off')
plt.subplot(224), plt.imshow(sobel_y, cmap='gray'), plt.title('Along Y-axis'), plt.axis('off')
plt.show()
```

**Q3) Write a python program to manipulate the excel sheet using openpyxl for the above application**

```python
# importing all the necessary libraries/modules
import openpyxl
from openpyxl.styles import Font, Alignment
```

```python
# instantiating an excel workbook
workbook = openpyxl.Workbook()
```

```python
# creating a user-defined function to update all the sheets

def updateSheet(sheet, title_, date):
    
    # changing the title to the one given in the parameter
    sheet.title = title_

    # setting the headers for the table
    sheet['B2'] = 'Fields'
    sheet['B2'].font = Font(bold = True) # styling
    sheet['B2'].alignment = Alignment(horizontal='center') # styling

    sheet['C2'] = 'Values'
    sheet['C2'].font = Font(bold = True) # styling
    sheet['C2'].alignment = Alignment(horizontal='center') # styling

    # field 1 - weather

    sheet['B3'] = 'Weather' # label
    sheet['B3'].alignment = Alignment(horizontal='center') # styling

    sheet['C3'] = weather_data[date]['Temperature'] # fetching value
    sheet['C3'].alignment = Alignment(horizontal='center') # styling

    # field 2 - humidity

    sheet['B4'] = 'Humidity' # label
    sheet['B4'].alignment = Alignment(horizontal='center') # styling

    sheet['C4'] = weather_data[date]['Humidity'] # fetching value
    sheet['C4'].alignment = Alignment(horizontal='center') # styling

    # field 3 - wind speed

    sheet['B5'] = 'Humidity' # label
    sheet['B5'].alignment = Alignment(horizontal='center') # styling

    sheet['C5'] = weather_data[date]['Humidity'] # fetching value
    sheet['C5'].alignment = Alignment(horizontal='center') # styling

    # field 4 - Rainfall

    sheet['B5'] = 'Rainfall' # label
    sheet['B5'].alignment = Alignment(horizontal='center') # styling

    sheet['C5'] = weather_data[date]['Rainfall'] # fetching value
    sheet['C5'].alignment = Alignment(horizontal='center') # styling

    # field 5 - Sunrise

    sheet['B6'] = 'Sunrise' # label
    sheet['B6'].alignment = Alignment(horizontal='center') # styling

    sheet['C6'] = weather_data[date]['Sunrise'] # fetching value
    sheet['C6'].alignment = Alignment(horizontal='center') # styling

    # field 5 - Sunset

    sheet['B7'] = 'Sunset' # label
    sheet['B7'].alignment = Alignment(horizontal='center') # styling

    sheet['C7'] = weather_data[date]['Sunset'] # fetching value
    sheet['C7'].alignment = Alignment(horizontal='center') # styling
```

```python
# updating first sheet
sheet1 = workbook.active
updateSheet(sheet1, "6th November", '06/11/23')
```

```python
# updating second sheet
sheet2 = workbook.create_sheet()
updateSheet(sheet2, "7th November", '07/11/23')
```
