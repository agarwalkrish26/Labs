## Q1) 1. Load the image, rename it

```python
# importing necessary libraries
import os
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the old and new file names
old_filename = 'D:/Z/Downloads/image17.jfif'
new_filename = 'D:/Z/Downloads/fruits.jpg'

# Checking if the old file exists 
if os.path.exists(old_filename):
    os.rename(old_filename, new_filename) # Rename the file
    print(f'Renamed {old_filename} to {new_filename}')
else:
    print(f'The file {old_filename} does not exist.')
```

```python
# importing the image
image = cv2.imread("D:/Z/Downloads/fruits.jpg")
```

```python
# Plotting the image
plt.title("Image")
plt.xlabel("X Scale")
plt.ylabel("Y Scale")
plt.imshow(image)
plt.show()
```

## Q2) Transform the image

```python
# changing the colour space to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# plotting the image
plt.title("Image")
plt.xlabel("X Scale")
plt.ylabel("Y Scale")
plt.imshow(image_rgb)
plt.show()
```

```python
kernel = np.array([[0, -1, 0],
                  [-1, 5, -1],
                  [0, -1, 0]])

image_sharpened = cv2.filter2D(image_rgb, -1, kernel)
```

```python
# plotting the image
plt.title("Scaled Image")
plt.xlabel("X Scale")
plt.ylabel("Y Scale")
plt.imshow(image_sharpened)
plt.show()
```

## Q3) Scale you image

```python
# Importing all the necessary libraries/modules
from PIL import Image
```

```python
image = Image.open('D:/Z/Downloads/fruits.jpg')
```

```python
rotated_45_image = image.rotate(45)
rotated_90_image = image.rotate(90)
rotated_180_image = image.rotate(180)
```

```python

fig1, ax = plt.subplots(1, 4, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image)
ax[0].set_title("Image")

ax[1].imshow(rotated_45_image)
ax[1].set_title("45D Rotated")

ax[2].imshow(rotated_90_image)
ax[2].set_title("90D Rotated")

ax[3].imshow(rotated_180_image)
ax[3].set_title("180D Rotated")

plt.show()
```

Splitting and merging

```python
R, G, B = image.split()
```

```python

plt.subplot(131),plt.imshow(R, cmap = 'gray'),plt.title('R Channel'), plt.axis('off')
plt.subplot(132),plt.imshow(G, cmap = 'gray'),plt.title('G Channel'),plt.axis('off')
plt.subplot(133),plt.imshow(B, cmap = 'gray'),plt.title('B Channel'),plt.axis('off')
plt.show()
```

```python
merged_image = Image.merge("RGB", (R, G, B))
display(merged_image)
```
