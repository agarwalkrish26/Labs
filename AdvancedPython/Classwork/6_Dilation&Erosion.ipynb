# `Dilation & Erosion`
<br>

## Basics:  
1. Erosion:
    - Erodes away the boundaries of the foreground object
    - Used to diminish the features of an image.
2. Dilation: 
    - Increases the object area
    - Used to accentuate features  
    
## Advantages:  

1. Erosion: 
    - It is useful for removing small white noises.
    - Used to detach two connected objects etc.
2. Dilation:
    - In cases like noise removal, erosion is followed by dilation. Because, erosion removes white noises, but it also shrinks our object. So we dilate it. Since noise is gone, they wonâ€™t come back, but our object area increases.
    - It is also useful in joining broken parts of an object.

```python
kernel = np.ones((3, 3), np.uint8)
  
image_erosion = cv2.erode(image, kernel, iterations=1)
image_dilation = cv2.dilate(image, kernel, iterations=1)
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image)
ax[0].set_title("Original Image")

ax[1].imshow(image_erosion)
ax[1].set_title("Eroded Image")

ax[2].imshow(image_dilation)
ax[2].set_title("Dilated Image")

plt.show()
```

# `Opening & Closing`  
<br>

## Basics:  
1. Opening:
    - Opening is a two-step operation that consists of erosion followed by dilation.  
    - It helps in removing small objects or noise while preserving the larger structures in an image. The process involves sliding a structuring element (a small binary matrix) over the image and performing erosion and dilation operations.

2. Closing: 
    - Closing is the opposite of opening 
    - Closing is a two-step operation performed by dilation followed by erosion. 
    - It helps in closing small holes or gaps in the foreground objects while preserving the overall shape and size of the objects.
    
## Advantages:  

1. Opening: 
    - removing salt-and-pepper noise
    - separating touching objects
    -  smoothing out the boundaries of objects
2. Closing:
    - filling small holes
    - connecting broken parts of objects
    -  smoothing out the boundaries of objects

```python
kernel = np.ones((5, 5), np.uint8)

image_opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
image_closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image)
ax[0].set_title("Original Image")

ax[1].imshow(image_opening)
ax[1].set_title("Opened Image")

ax[2].imshow(image_closing)
ax[2].set_title("Closed Image")

plt.show()
```
