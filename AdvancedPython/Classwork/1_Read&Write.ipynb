### Printing a picture

```python
import matplotlib.pyplot as plt
```

```python
# uploading the image
file = 'D:/Z/Downloads/Lenna_(test_image).png'
image = cv2.imread(file)
```

```python
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # RGB Format
image_GRAY = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # GRAYSCALE Format
image_HSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) # HSV Format
image_YUV = cv2.cvtColor(image, cv2.COLOR_BGR2YUV) # YUV Format
```

### Subplotting the images

**Method 1**

```python
plt.subplot(221),plt.imshow(image_RGB, cmap = 'gray'),plt.title('RGB Image'), plt.axis('off')
plt.subplot(222),plt.imshow(image_GRAY, cmap = 'gray'),plt.title('GRAY Image'),plt.axis('off')
plt.subplot(223),plt.imshow(image_HSV, cmap = 'gray'),plt.title('HSV Image'),plt.axis('off')
plt.subplot(224),plt.imshow(image_YUV, cmap = 'gray'),plt.title('YUV Image'),plt.axis('off')
plt.show()
```

**Method 2**

```python
fig1, ax = plt.subplots(1, 4, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image_RGB)
ax[0].set_title("RGB Image")

ax[1].imshow(image_GRAY)
ax[1].set_title("GRAY Image")

ax[2].imshow(image_HSV)
ax[2].set_title("HSV Image")

ax[3].imshow(image_YUV)
ax[3].set_title("YUV Image")

plt.show()
```
