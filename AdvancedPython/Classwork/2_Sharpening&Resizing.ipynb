# `Reading and Writing an Image`

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# uploading the image
file = 'D:/Z/Downloads/Lenna_(test_image).png'
image = cv2.imread(file)
```

```python
# RGB Format

image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.title("Lenna - RGB")
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
 
plt.imshow(image_RGB) 
plt.show()
```

# `Sharpening`

### Kernal based filtering

```python
import numpy as np
```

```python
kernel = np.array([[0, -1, 0],
                  [-1, 5, -1],
                  [0, -1, 0]])

image_sharpened = cv2.filter2D(image_RGB, -1, kernel)
image_sharpened_2 = cv2.filter2D(image_RGB, -29, kernel)
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image_RGB)
ax[0].set_title("RGB Image")

ax[1].imshow(image_sharpened)
ax[1].set_title("Sharpened Image")

ax[2].imshow(image_sharpened_2)
ax[2].set_title("Sharpened_2 Image")

plt.show()
```

### Laplacian Filtering

```python
image_sharpened_3 = cv2.Laplacian(image, cv2.CV_64F)
image_sharpened_4 = cv2.Laplacian(image, cv2.CV_32F)
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image_RGB)
ax[0].set_title("RGB Image")

ax[1].imshow(image_sharpened_3)
ax[1].set_title("Laplacian 64F Image")

ax[2].imshow(image_sharpened_4)
ax[2].set_title("Laplacian 32F Image")

plt.show()
```

# `Resizing`

```python
new_size = (300, 300)
image_resized = cv2.resize(image_RGB, new_size)
```

```python
fig1, ax = plt.subplots(1, 2, figsize = (6, 6))
fig1.tight_layout()

ax[0].imshow(image_resized)
ax[0].set_title("Original Image")

ax[1].imshow(image)
ax[1].set_title("Resized Image")

plt.show()
```
