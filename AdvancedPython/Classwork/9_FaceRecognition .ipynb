```python
# Importing all the necessary libraries
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

### Method 1: Detecting face via a given input

```python
face_cascade = cv2.CascadeClassifier('D:/Z/Downloads/haarcascade_frontalface_default.xml')
```

```python
# For an image
image = cv2.imread('/Users/shreyansh/Desktop/Krish/SEM5/christianbale.png')
```

```python
# For an image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Method 2: Detecting face via live camera feed

```python
face_cascade = cv2.CascadeClassifier('D:/Z/Downloads/haarcascade_frontalface_default.xml')
```

```python
video_capture = cv2.VideoCapture(0)  # 0 for the default camera
```

```python
while True:
    ret, frame = video_capture.read()
    
    if not ret:
        break
    
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
    cv2.imshow('Face Recognition', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```
