```python
# Importing all the necessary libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# uploading the image
image = cv2.imread('D:/Z/Downloads/Lenna_(test_image).png')

# RGB Format
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# GRAYSCALE Format
image_GRAY = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image)
ax[0].set_title("Original Image")

ax[1].imshow(image_RGB)
ax[1].set_title("RGB Image")

ax[2].imshow(image_GRAY)
ax[2].set_title("GRAYSCALE Image")

plt.show()
```

# `Sobel Method`

### 1.1) RGB Image

```python
sobelx = cv2.Sobel(image_RGB, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=7) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(image_RGB, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=7) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(image_RGB, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=7) # Combined X and Y Sobel Edge Detection
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(sobelx)
ax[0].set_title("X axis Edges")

ax[1].imshow(sobely)
ax[1].set_title("Y axis Edges")

ax[2].imshow(sobelxy)
ax[2].set_title("Combined axis Edges")

plt.show()
```

### 1.2) RGB Image with GaussianBlur

```python
# Blurred image
image_blurred = cv2.GaussianBlur(image_RGB, (3, 3), 0) 

sobelx_ = cv2.Sobel(image_blurred, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=7) # Sobel Edge Detection on the X axis
sobely_ = cv2.Sobel(image_blurred, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=7) # Sobel Edge Detection on the Y axis
sobelxy_ = cv2.Sobel(image_blurred, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=7) # Combined X and Y Sobel Edge Detection
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(sobelx_)
ax[0].set_title("X axis Edges")

ax[1].imshow(sobely_)
ax[1].set_title("Y axis Edges")

ax[2].imshow(sobelxy_)
ax[2].set_title("Combined axis Edges")

plt.show()
```

### 2.1) Grayscale Image

```python
sobelx_G = cv2.Sobel(image_GRAY, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=7) # Sobel Edge Detection on the X axis
sobely_G = cv2.Sobel(image_GRAY, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=7) # Sobel Edge Detection on the Y axis
sobelxy_G = cv2.Sobel(image_GRAY, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=7) # Combined X and Y Sobel Edge Detection
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(sobelx_G)
ax[0].set_title("X axis Edges")

ax[1].imshow(sobely_G)
ax[1].set_title("Y axis Edges")

ax[2].imshow(sobelxy_G)
ax[2].set_title("Combined axis Edges")

plt.show()
```

### 2.2) Grayscale Image with GaussianBlur

```python
# Blurred image
image_blurred_G = cv2.GaussianBlur(image_GRAY, (3, 3), 0) 

sobelx_G_ = cv2.Sobel(image_blurred_G, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=7) # Sobel Edge Detection on the X axis
sobely_G_ = cv2.Sobel(image_blurred_G, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=7) # Sobel Edge Detection on the Y axis
sobelxy_G_ = cv2.Sobel(image_blurred_G, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=7) # Combined X and Y Sobel Edge Detection
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(sobelx_G_)
ax[0].set_title("X axis Edges")

ax[1].imshow(sobely_G_)
ax[1].set_title("Y axis Edges")

ax[2].imshow(sobelxy_G_)
ax[2].set_title("Combined axis Edges")

plt.show()
```

# `Laplacian Method`

### 1.1) RGB Image

```python
image_laplacian_64 = cv2.Laplacian(image_RGB, cv2.CV_64F)
image_laplacian_32 = cv2.Laplacian(image_RGB, cv2.CV_32F)
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image_RGB)
ax[0].set_title("RGB Image")

ax[1].imshow(image_laplacian_64)
ax[1].set_title("Laplacian 64F Image")

ax[2].imshow(image_laplacian_32)
ax[2].set_title("Laplacian 32F Image")

plt.show()
```

### 1.2) RGB Image with GaussianBlur

```python
image_laplacian_64_ = cv2.Laplacian(image_blurred, cv2.CV_64F)
image_laplacian_32_ = cv2.Laplacian(image_blurred, cv2.CV_32F)
```
