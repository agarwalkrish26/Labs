# `Perspective Transformation`  
<br>

1. For getting better information about an image, we can change the perspective of a video or an image.
1. We use two functions for Perspective transformation 
    - getPerspectiveTransform() 
    - warpPerspective()
    
    
### **Code Explanation**

1. points_A and points_B: 
    - These points represent corresponding corners in two different coordinate systems. 
    - points_A represents the corners in the source image
    - points_B represents the corners in the desired output image
    - Each point is defined as (x, y) coordinates

1. getPerspectiveTransform function 
    - called with points_A and points_B as arguments
    - This function calculates the perspective transformation matrix (M) that maps points from points_A to points_B.

1. The warpPerspective function
    - called with the original image (image), the perspective transformation matrix (M), and the output image size (specified as (420, 594))
    - This function applies the perspective transformation to the input image and generates the output image (warped).

1. Overall, this code performs a perspective transformation on an image, warps it according to the defined source and destination points, and displays the original and transformed images side by side for visual comparison.

```python
# Defining the source and output points
points_A = np.float32([[320,50], [600,215], [85,610], [530,780]])
points_B = np.float32([[0,0], [420,0], [0,594], [420,594]])

# Defining perspective transformers
M = cv2.getPerspectiveTransform(points_A, points_B)
warped = cv2.warpPerspective(image_RGB, M, (420,594))

# sub-plotting the images
plt.figure(figsize=(20, 20))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_RGB)

plt.subplot(1, 2, 2)
plt.title("warpPerspective")
plt.imshow(warped)
plt.show()
```

# `Image Pyramid`  
<br>

1. A very useful technique when we required scaling in object detection. 
1. Two common kinds of image pyramids 
    - Gaussian
    - Laplacian pyramid
1. Using the pyrUp() function in OpenCV to downsample an image.
1. Using the pyrDown() function in OpenCV to upsample an image.

```python
smaller = cv2.pyrDown(image_RGB)
larger = cv2.pyrUp(smaller)

# sub-plotting the original, pyramid downed and pyramid up images
plt.figure(figsize=(20, 20))

plt.subplot(2, 2, 1)
plt.title("Original")
plt.imshow(image_RGB)

plt.subplot(2, 2, 2)
plt.title("Smaller")
plt.imshow(smaller)

plt.subplot(2, 2, 3)
plt.title("Larger")
plt.imshow(larger)
plt.show()
```
