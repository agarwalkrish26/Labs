```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("D:/Z/Downloads/Lenna_(test_image).png")
```

```python
# Sharpened RGB image
kernel = np.array([[0, -1, 0],
                  [-1, 5, -1],
                  [0, -1, 0]])

image_RGB = cv2.filter2D(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), -1, kernel)
```

```python
plt.title("Lenna - RGB")
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
 
plt.imshow(image_RGB) 
plt.show()
```

# Plotting only 1 component of an image

```python
# Sub plot of multiple images

fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image_RGB[:, :, 0])
ax[0].set_title("Red Component")

ax[1].imshow(image_RGB[:, :, 1])
ax[1].set_title("Green Component")

ax[2].imshow(image_RGB[:, :, 2])
ax[2].set_title("Blue Component")

plt.show()
```

# Masking: Chaning the background of an image

### Method 1.1: Solid Background

##### RGB Image

```python
# Importing an image with object 
image2 = cv2.imread("D:/Z/Downloads/pizza.jpg")
image2_copy = np.copy(image2)

# RGB Image
image2_RGB = cv2.cvtColor(image2_copy, cv2.COLOR_BGR2RGB)
```

```python
# Setting up the lower and upper values of the mask
lower_blue = np.array([0, 150, 220])
upper_blue = np.array([255, 255, 255])

# Creating a mask of an image
mask = cv2.inRange(image2_RGB, lower_blue, upper_blue)

# Creating a new image with layer over layer
masked_image = np.copy(image2_RGB)
masked_image[mask != 0] = [0, 255, 0] # [Red, Green, Blue]
```

```python
# Sub plot of multiple images

fig1, ax = plt.subplots(1, 4, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image2_copy)
ax[0].set_title("Original Image")

ax[1].imshow(image2_RGB)
ax[1].set_title("RGB Image")

ax[2].imshow(mask)
ax[2].set_title("Mask Layer")

ax[3].imshow(masked_image)
ax[3].set_title("Final Image")

plt.show()
```

##### Grayscale Image

```python
# Grayscale Image
image2_GRAY = cv2.cvtColor(image2_copy, cv2.COLOR_BGR2GRAY)

# Setting up the lower and upper values of the mask
lower_blue2 = 250
upper_blue2 = 255
```

```python
# Creating a mask of an image
mask2 = cv2.inRange(image2_GRAY, lower_blue2, upper_blue2)

# Creating a new image with layer over layer
masked_image2 = np.copy(image2_RGB)
masked_image2[mask2 != 0] = [0, 255, 0]
```

```python
# Sub plot of multiple images

fig1, ax = plt.subplots(1, 4, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image2)
ax[0].set_title("Original Image")

ax[1].imshow(image2_GRAY)
ax[1].set_title("GRAYSCALE Image")

ax[2].imshow(mask2)
ax[2].set_title("Mask Layer")

ax[3].imshow(masked_image2)
ax[3].set_title("Final Image")

plt.show()
```
