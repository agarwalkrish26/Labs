```python
# Importing all the necessary libraries/modules
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image = Image.open('D:/Z/Downloads/Lenna_(test_image).png')
display(image)
```

# `Rotating an Image`

```python
rotated_45_image = image.rotate(45)
rotated_90_image = image.rotate(90)
rotated_180_image = image.rotate(180)
```

```python
fig1, ax = plt.subplots(1, 4, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image)
ax[0].set_title("Image")

ax[1].imshow(rotated_45_image)
ax[1].set_title("45D Rotated")

ax[2].imshow(rotated_90_image)
ax[2].set_title("90D Rotated")

ax[3].imshow(rotated_180_image)
ax[3].set_title("180D Rotated")

plt.show()
```

# `Split and Merge`

## Split

```python
R, G, B = image.split()
```

```python
plt.subplot(131),plt.imshow(R, cmap = 'gray'),plt.title('R Channel'), plt.axis('off')
plt.subplot(132),plt.imshow(G, cmap = 'gray'),plt.title('G Channel'),plt.axis('off')
plt.subplot(133),plt.imshow(B, cmap = 'gray'),plt.title('B Channel'),plt.axis('off')
plt.show()
```

## Merge

**Merging all the colour channels back together**

```python
merged_image = Image.merge("RGB", (R, G, B))
display(merged_image)
```

**Merrging 2 images side-by-side**

**`merged_image_2 = Image.new('RGB', (2 * image.size[0], image.size[1]), (250, 250, 250))`**  
This line creates a new image object called merged_image_2 using the Image.new() function. The new image has the mode 'RGB', a width twice the original image's width (2 * image.size[0]), and the same height as the original image (image.size[1]). The (250, 250, 250) argument sets the initial color of the image to a light gray RGB value.  
<br>
**`merged_image_2.paste(image, (0, 0))`**  
**`merged_image_2.paste(rotated_180_image, (image.size[0], 0))`**  
These lines paste the original image and the rotated image onto the merged_image_2 side by side. The first paste() call pastes the image at the top-left corner of merged_image_2 (position (0, 0)), and the second paste() call pastes the rotated_180_image next to the original image, starting at the right edge of the original image (position (image.size[0], 0)).

```python
merged_image_2 = Image.new('RGB', (2 * image.size[0], image.size[1]), (250, 250, 250))
merged_image_2.paste(image, (0, 0))
merged_image_2.paste(rotated_180_image, (image.size[0], 0))
```
