### **Q2) Convert it into various color models; Display the images in tabular form**

- RGB (Red, Green, Blue):

    - RGB is the most common color space used in digital displays and imaging devices.
    - It represents colors as a combination of red, green, and blue primary colors.
    - Each pixel is defined by the intensity values of these three primary colors, typically ranging from 0 to 255.
    - By varying the intensity of each primary color, a wide range of colors can be represented.


- Grayscale:

    - Grayscale is a color space where each pixel is represented by a single intensity value ranging from 0 to 255.
    - It is often used in applications where color information is not necessary or to simplify image processing.
    - In grayscale, the absence of color is represented as varying shades of gray.
    - It is commonly used for black-and-white images and to perform certain image operations like edge detection and thresholding.


- HSV (Hue, Saturation, Value):

    - HSV is a color space that separates the color information into three components: hue, saturation, and value.
    - Hue represents the color itself, such as red, blue, or yellow, and is expressed as an angle ranging from 0 to 360 degrees.
    - Saturation refers to the purity or intensity of the color, with lower values indicating more muted colors and higher values indicating more vivid colors.
    - Value represents the brightness or lightness of the color, ranging from black (0) to white (maximum value).
    - HSV is often used in computer vision tasks like color segmentation and object tracking.


- YUV:

    - YUV is a color space used in video and image processing to separate color information from luminance (brightness) information.
    - Y represents the luma component, which represents the brightness of the image.
    - U and V are the chroma components, representing the color information.
    - Separating luminance and chrominance allows for more efficient image and video compression.
    - YUV is commonly used in video encoding formats like MPEG and JPEG.


- Binary:

    - Binary is not a specific color space but refers to a representation where each pixel is either black (0) or white (1).
    - It is often used for binary image processing operations such as thresholding and image segmentation.
    - In a binary image, pixels are classified into foreground (object) or background based on a certain threshold.


- LAB:

    - LAB is a color space that represents colors in a device-independent manner.
    - L represents the lightness component ranging from black (0) to white (maximum value).
    - A and B are the color-opponent dimensions, where A represents the green-red axis and B represents the blue-yellow axis.
    - LAB is often used in color-related applications, such as color correction, color difference calculations, and perceptual color models.


- YCrCb (Y'CbCr):

    - YCrCb is a color space used in digital imaging and video systems.
    - Y represents the luma component, similar to YUV.
    - Cr and Cb are the chroma components, representing the color difference between red and luma and blue and luma, respectively.
    - YCrCb is commonly used in image and video compression formats like JPEG and MPEG.
    - It provides a more efficient representation of color information compared to RGB, especially for human visual perception.

```python
# Converting the original image to different colour spaces

image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # RGB Image
image_GRAY = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #GRAYSCALE Image
image_HSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) # HSV Image
image_YUV = cv2.cvtColor(image, cv2.COLOR_BGR2YUV) # YUV Image
bret, image_BIN = cv2.threshold(image_GRAY, 127, 255, cv2.THRESH_BINARY) # BINARY Image
image_INV_BIN = cv2.bitwise_not(image_BIN) # INVERSE BINARY Image
image_LAB = cv2.cvtColor(image, cv2.COLOR_BGR2LAB) # LAB Image
image_YCrCb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb) # YCrCb Image
```

##### Plotting 1

```python
plt.subplot(241),plt.imshow(image_RGB, cmap = 'gray'),plt.title('RGB'), plt.axis('off')
plt.subplot(242),plt.imshow(image_GRAY, cmap = 'gray'),plt.title('GRAY'),plt.axis('off')
plt.subplot(243),plt.imshow(image_HSV, cmap = 'gray'),plt.title('HSV'),plt.axis('off')
plt.subplot(244),plt.imshow(image_YUV, cmap = 'gray'),plt.title('YUV'),plt.axis('off')
plt.subplot(245),plt.imshow(image_BIN, cmap = 'gray'),plt.title('BINARY'),plt.axis('off')
plt.subplot(246),plt.imshow(image_INV_BIN, cmap = 'gray'),plt.title('INVERSE BINARY'),plt.axis('off')
plt.subplot(247),plt.imshow(image_LAB, cmap = 'gray'),plt.title('LAB'),plt.axis('off')
plt.subplot(248),plt.imshow(image_YCrCb, cmap = 'gray'),plt.title('YCrCb'),plt.axis('off')
plt.show()
```

#### Plotting 2

```python
fig1, ax = plt.subplots(2, 4, figsize = (8, 8))
fig1.tight_layout()

ax[0, 0].imshow(image_RGB)
ax[0, 0].set_title("RGB")

ax[0, 1].imshow(image_GRAY)
ax[0, 1].set_title("GRAY")

ax[0, 2].imshow(image_HSV)
ax[0, 2].set_title("HSV")

ax[0, 3].imshow(image_YUV)
ax[0, 3].set_title("YUV")

ax[1, 0].imshow(image_BIN)
ax[1, 0].set_title("BINARY")

ax[1, 1].imshow(image_INV_BIN)
ax[1, 1].set_title("INVERSE BINARY")

ax[1, 2].imshow(image_LAB)
ax[1, 2].set_title("LAB")

ax[1, 3].imshow(image_YCrCb)
ax[1, 3].set_title("YCrCb")

plt.show()
```
