```python
# importing necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# importing the image
image = cv2.imread("D:/Z/Downloads/shapes.png")

# converting it to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
fig1, ax = plt.subplots(1, 2, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image)
ax[0].set_title("Image")

ax[1].imshow(image_rgb)
ax[1].set_title("RGB Image")

plt.show()
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define the parameter for the Harris Corner Detector
block_size = 2  # Size of the neighborhood considered for corner detection
ksize = 3       # Aperture parameter for the Sobel derivative operator
k = 0.04        # Harris detector free parameter, usually in the range [0.04, 0.06]

# Apply Harris Corner Detection
corner_harris = cv2.cornerHarris(gray_image, block_size, ksize, k)

# Threshold to identify corners
threshold = 0.06  # Adjust this threshold as needed
corner_harris[corner_harris > threshold * corner_harris.max()] = 255

# Create an output image to mark the corners
output_image = np.copy(image)
output_image[corner_harris > 0] = [0, 255, 255]  # Mark detected corners as red

# Display the original image with detected corners
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
plt.title('Harris Corner Detection')
plt.axis('off')
plt.show()
```
