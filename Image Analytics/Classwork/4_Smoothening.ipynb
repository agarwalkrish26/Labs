```python
# importing necessary libraries
import cv2 
import matplotlib.pyplot as plt
import pydicom
import numpy as np
```

```python
# importig the image through the pydicom library
original_image = pydicom.read_file("D:/Z/Downloads/ID_000012eaf.dcm")

# converting the image to a matrix
image = original_image.pixel_array
```

```python
plt.xlabel("X-axis Scale")
plt.ylabel("Y-axis Scale")
plt.title("Original Image")
plt.imshow(image)
plt.show()
```

## Smoothening

```python
# Applying gaussian blur to the image
k_size = (5, 5) # kernel size
sigma_x = 0

kernel = np.ones((5, 5), np.float32)/25
```

```python
gaussian_image = cv2.GaussianBlur(image, k_size, sigma_x)
average_image = cv2.blur(image, k_size)
medain_image = cv2.medianBlur(image, 5)
# bilateral_image = cv2.bilateralFilter(image, 9, 75, 75)
image_2d = cv2.filter2D(src = image, ddepth = -1, kernel = kernel)
```

```python
plt.subplot(231),plt.imshow(image, cmap = 'gray'),plt.title('Original'), plt.axis('off')
plt.subplot(232),plt.imshow(gaussian_image, cmap = 'gray'),plt.title('Gaussian'), plt.axis('off')
plt.subplot(233),plt.imshow(average_image, cmap = 'gray'),plt.title('Average'), plt.axis('off')
plt.subplot(234),plt.imshow(medain_image, cmap = 'gray'),plt.title('Median'), plt.axis('off')
plt.subplot(235),plt.imshow(medain_image, cmap = 'gray'),plt.title('Median'), plt.axis('off')
plt.plot()
```
