```python
# Importing all the necessary libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# uploading the image
image = cv2.imread('D:/Z/Downloads/Lenna_(test_image).png')

# RGB Format
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# GRAYSCALE Format
image_GRAY = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image)
ax[0].set_title("Original Image")

ax[1].imshow(image_RGB)
ax[1].set_title("RGB Image")

ax[2].imshow(image_GRAY)
ax[2].set_title("GRAYSCALE Image")

plt.show()
```

# `Canny Method`

**1.1) RGB Image**

```python
image_canny = cv2.Canny(image_RGB, 100, 150)
```

```python
fig1, ax = plt.subplots(1, 2, figsize = (6, 6))
fig1.tight_layout()

ax[0].imshow(image_RGB)
ax[0].set_title("Image")

ax[1].imshow(image_canny)
ax[1].set_title("Edges")

plt.show()
```

**1.2) RGB Image with GaussianBlur**

```python
# Blurred image
image_blurred = cv2.GaussianBlur(image_RGB, (3, 3), 0) 

image_canny_ = cv2.Canny(image_blurred, 100, 102)
```

```python
fig1, ax = plt.subplots(1, 2, figsize = (6, 6))
fig1.tight_layout()

ax[0].imshow(image_RGB)
ax[0].set_title("Image")

ax[1].imshow(image_canny_)
ax[1].set_title("Edges")

plt.show()
```

**2.1) Grayscale Image**

```python
image_canny_g = cv2.Canny(image_GRAY, 100, 200)
```

```python
fig1, ax = plt.subplots(1, 2, figsize = (6, 6))
fig1.tight_layout()

ax[0].imshow(image_GRAY)
ax[0].set_title("Image")

ax[1].imshow(image_canny_g)
ax[1].set_title("Edges")

plt.show()
```

**2.2) Grayscale Image with GaussianBlur**

```python
# Blurred image
image_blurred_G = cv2.GaussianBlur(image_GRAY, (3, 3), 0) 

image_canny_g_ = cv2.Canny(image_blurred_G, 100, 102)
```

```python
fig1, ax = plt.subplots(1, 2, figsize = (6, 6))
fig1.tight_layout()

ax[0].imshow(image_blurred_G)
ax[0].set_title("Image")

ax[1].imshow(image_canny_g_)
ax[1].set_title("Edges")

plt.show()
```

# `Contour`

```python
# uploading the image
image_ = cv2.imread('D:/Z/Downloads/pizza.jpg')

# RGB Format
image_RGB_ = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# GRAYSCALE Format
image_GRAY_ = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image_)
ax[0].set_title("Original Image")

ax[1].imshow(image_RGB_)
ax[1].set_title("RGB Image")

ax[2].imshow(image_GRAY_)
ax[2].set_title("GRAYSCALE Image")

plt.show()
```

**1.1) RGB Image**

```python
# Find Canny edges
edged = cv2.Canny(image_GRAY_, 30, 200)
edged_ = edged.copy()
  
# Finding Contours
# Use a copy of the image e.g. edged.copy() since findContours alters the image
contours, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    
# Draw all contours, -1 signifies drawing all contours
cv2.drawContours(image_RGB_, contours, -1, (0, 0, 255), 3)
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(edged_)
ax[0].set_title("Canny Image")

ax[1].imshow(edged)
ax[1].set_title("Contoured Image")

ax[2].imshow(image_)
ax[2].set_title("Canny+Contoured Image")

plt.show()
```
