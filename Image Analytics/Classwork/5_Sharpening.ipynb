```python
# importing necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# importing the image
image = cv2.imread("D:/Z/Downloads/Lenna_(test_image).png")

# converting it to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
fig1, ax = plt.subplots(1, 2, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image)
ax[0].set_title("Image")

ax[1].imshow(image_rgb)
ax[1].set_title("RGB Image")

plt.show()
```

## Sharpening

**Kernel based Filterin**

```python
kernel1 = np.array([[0, -1, 0],
                  [-1, 5, -1],
                  [0, -1, 0]])

image_sharpened1 = cv2.filter2D(image_rgb, -1, kernel1)
```

```python
fig1, ax = plt.subplots(1, 2, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image_rgb)
ax[0].set_title("Image")

ax[1].imshow(image_sharpened1)
ax[1].set_title("Sharpened Image")

plt.show()
```

**Laplacian Filtering**

```python
image_sharpened2 = cv2.Laplacian(image, cv2.CV_64F)
image_sharpened3 = cv2.Laplacian(image, cv2.CV_32F)
image_sharpened4 = cv2.Laplacian(image, cv2.CV_8U)
```

```python
fig1, ax = plt.subplots(2, 2, figsize = (5, 5))
fig1.tight_layout()

ax[0, 0].imshow(image_rgb)
ax[0, 0].set_title("Image")

ax[0, 1].imshow(image_sharpened4)
ax[0, 1].set_title("Laplacian 8U Image")

ax[1, 0].imshow(image_sharpened2)
ax[1, 0].set_title("Laplacian 64F Image")

ax[1, 1].imshow(image_sharpened3)
ax[1, 1].set_title("Laplacian 32F Image")

plt.show()
```

**Sobel Method**

```python
sobelx = cv2.Sobel(image_rgb, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(image_rgb, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(image_rgb, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3) # Combined X and Y Sobel Edge Detection
```

```python
fig1, ax = plt.subplots(1, 3, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(sobelx)
ax[0].set_title("X axis Edges")

ax[1].imshow(sobely)
ax[1].set_title("Y axis Edges")

ax[2].imshow(sobelxy)
ax[2].set_title("Combined axis Edges")

plt.show()
```
