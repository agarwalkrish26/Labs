# `Class 1: Read and Write`

```python
# importing all the necessary libraries/moduels
import cv2
import matplotlib.pyplot as plt
```

```python
# uploading the image
file = 'D:/Z/Downloads/Lenna_(test_image).png'
image = cv2.imread(file)
```

```python
# RGB Format

image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.title("Lenna - RGB")
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
 
plt.imshow(image_RGB) 
plt.show()
```

```python
# GRAYSCALE Format

image_GRAY = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.title("Lenna - GRAY")
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
 
plt.imshow(image_GRAY)
plt.show()
```

```python
# HSV Format

image_HSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

plt.title("Lenna - HSV")
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
 
plt.imshow(image_HSV) 
plt.show()
```

```python
# YUV Format

image_YUV = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)

plt.title("Lenna - YUV")
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
 
plt.imshow(image_YUV) 
plt.show()
```

```python
# PLottig all images together in a subplot

fig1, ax = plt.subplots(1, 4, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image_RGB)
ax[0].set_title("RGB Image")

ax[1].imshow(image_GRAY)
ax[1].set_title("GRAY Image")

ax[2].imshow(image_HSV)
ax[2].set_title("HSV Image")

ax[3].imshow(image_YUV)
ax[3].set_title("YUV Image")
```

# `Class 2: Binary Image`

```python
# BINARY IMAGES

bret, image_BIN = cv2.threshold(image_GRAY, 127, 255, cv2.THRESH_BINARY)

plt.title("Lenna - BINARY")
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
 
plt.imshow(image_BIN) 
plt.show()
```

```python
image_INV_BIN = cv2.bitwise_not(image_BIN)

plt.title("Lenna - INVERSE BINARY")
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
 
plt.imshow(image_INV_BIN) 
plt.show()
```

```python
# PLottig all images together in a subplot

fig1, ax = plt.subplots(1, 2, figsize = (4, 4))
fig1.tight_layout()

ax[0].imshow(image_BIN)
ax[0].set_title("Binary Image")

ax[1].imshow(image_INV_BIN)
ax[1].set_title("Inverse Binary Image")
```
