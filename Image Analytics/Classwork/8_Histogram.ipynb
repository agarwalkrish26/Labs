```python
# importing necessary libraries
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# importing the image to RGB
image = cv2.cvtColor(cv2.imread("D:/Z/Downloads/dollars.jpeg"), cv2.COLOR_BGR2RGB)

# converting the image to Grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# plotting the image
plt.title("Dollars")
plt.imshow(image)
plt.show()
```

## `Histogram Equalization`

**GRAYSCALE Image**

```python
# Calculate the histogram of the original image
hist_original = cv2.calcHist([image_gray], [0], None, [256], [0, 256])

# Perform histogram equalization
equalized_image = cv2.equalizeHist(image_gray)

# Calculate the histogram of the equalized image
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
```

```python
# Display the original and equalized images side by side
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.imshow(image_gray, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')

# Display histograms
plt.subplot(2, 2, 3)
plt.plot(hist_original, color='black')
plt.title('Original Image Histogram')
plt.xlim([0, 256])

plt.subplot(2, 2, 4)
plt.plot(hist_equalized, color='black')
plt.title('Equalized Image Histogram')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

**RGB Image**

```python
# Split the color image into its RGB channels
blue_channel, green_channel, red_channel = cv2.split(image)

# Calculate the histograms for each channel of the original image
hist_blue = cv2.calcHist([blue_channel], [0], None, [256], [0, 256])
hist_green = cv2.calcHist([green_channel], [0], None, [256], [0, 256])
hist_red = cv2.calcHist([red_channel], [0], None, [256], [0, 256])

# Equalize the color image
equalized_image = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
equalized_image[:, :, 0] = cv2.equalizeHist(equalized_image[:, :, 0])
equalized_image = cv2.cvtColor(equalized_image, cv2.COLOR_YCrCb2BGR)

# Split the equalized color image into its RGB channels
equalized_blue_channel, equalized_green_channel, equalized_red_channel = cv2.split(equalized_image)

# Calculate the histograms for each channel of the equalized image
hist_equalized_blue = cv2.calcHist([equalized_blue_channel], [0], None, [256], [0, 256])
hist_equalized_green = cv2.calcHist([equalized_green_channel], [0], None, [256], [0, 256])
hist_equalized_red = cv2.calcHist([equalized_red_channel], [0], None, [256], [0, 256])
```

```python
# Display the original and equalized images and histograms side by side
plt.figure(figsize=(15, 12))

# Show the original color image
plt.subplot(2, 4, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Color Image')

# Show the histograms for each channel of the original image
plt.subplot(2, 4, 2)
plt.plot(hist_blue, color='blue')
plt.title('Blue Channel Histogram (Original)')
plt.xlim([0, 256])

plt.subplot(2, 4, 3)
plt.plot(hist_green, color='green')
plt.title('Green Channel Histogram (Original)')
plt.xlim([0, 256])

plt.subplot(2, 4, 4)
plt.plot(hist_red, color='red')
plt.title('Red Channel Histogram (Original)')
plt.xlim([0, 256])

# Show the equalized color image
plt.subplot(2, 4, 5)
plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB))
plt.title('Equalized Color Image')

# Show the histograms for each channel of the equalized image
plt.subplot(2, 4, 6)
plt.plot(hist_equalized_blue, color='blue')
plt.title('Blue Channel Histogram (Equalized)')
plt.xlim([0, 256])

plt.subplot(2, 4, 7)
plt.plot(hist_equalized_green, color='green')
plt.title('Green Channel Histogram (Equalized)')
plt.xlim([0, 256])

plt.subplot(2, 4, 8)
plt.plot(hist_equalized_red, color='red')
plt.title('Red Channel Histogram (Equalized)')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
