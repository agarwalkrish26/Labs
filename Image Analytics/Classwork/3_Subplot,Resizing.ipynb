```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# `Sub-Plotting`

```python
# Setting a seed value 
np.random.seed(34)

# X-coordinates
x = np.linspace(0, 1000, 20)

# Y-coordinates
y1 = np.sin(x)
y2 = x**3/2
y3 = x**2
y4 = x**5/2
```

```python
# Create a grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot each array in its respective subplot
axs[0, 0].plot(x, y1, color='red')
axs[0, 0].set_title('x vs y1')

axs[0, 1].plot(x, y2, color='blue')
axs[0, 1].set_title('x vs y2')

axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('x vs y3')

axs[1, 1].plot(x, y4, color='magenta')
axs[1, 1].set_title('x vs y4')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

# `Resizing`

```python
# uploading an image
image = cv2.imread('D:/Z/Downloads/Lenna_(test_image).png')

# converting the image to RGB
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# PLotting the RGB formatted image
plt.title("Lenna - RGB")
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
plt.imshow(image_RGB) 
plt.show()
```

```python
print("Inital Dimensions of the image:", image_RGB.shape)
```

**Resizing**

```python
# defining the new size
new_size = (300, 300)

# resizing the image
image_resized = cv2.resize(image_RGB, new_size)
```

```python
# plotting the initial and the resized images on axis

fig1, ax = plt.subplots(1, 2, figsize = (8, 8))
fig1.tight_layout()

ax[0].imshow(image_RGB)
ax[0].set_title("Original Image")

ax[1].imshow(image_resized)
ax[1].set_title("Resized Image")

plt.plot()
```

```python
print("New Dimensions of the image:", image_resized.shape)
```

### Trying out various parameters inside .resize

```python
half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv2.resize(image, (1050, 1610))
 
stretch_near = cv2.resize(image, (780, 540),
               interpolation = cv2.INTER_LINEAR)
 
 
Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = 4
 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
 
plt.show()
```
