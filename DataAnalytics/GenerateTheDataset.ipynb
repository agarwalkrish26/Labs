```python
# Importing all the necesasry modules
import pandas as pd
import numpy as np
import random
from random import randint
from random import choice
```

#### DF1

```python
# ------------------ Q1) Take a domain on your own ------------------

min_emp, max_emp = 1000, 3000
designation = ['Senior Prof', 'Junior Prof', 'IT', 'Staff']
min_age, max_age = 25, 60
gender = ['Male', 'Female', 'Others']
min_salary, max_salary = 30000, 200000
min_year, max_year = 0, 19

df = pd.DataFrame(columns = ['EMP ID', 'Designation', 'Age', 'Gender', 'Salary', 'Experience (in yrs)'])
```

```python
print("-"*15)
count = int(input("Number of rows: "))
print("-"*15, "\n")

random.seed(69)

for c in range(0, count):
    new_entry = {}
    new_entry['EMP ID'] = randint(min_emp, max_emp)
    new_entry['Designation'] = random.choice(designation)
    new_entry['Age'] = randint(min_age, max_age)
    new_entry['Gender'] = random.choice(gender)
    new_entry['Salary'] = randint(min_salary, max_salary)
    new_entry['Experience'] = randint(min_year, max_year)
    df = df.append(new_entry, ignore_index = True)

df
```

**Observation:** When we created the dataframe initially, we made column named 'Experience (in yrs)' and while appending with the dictionary, we created 'Experience'. So, it adds a new columnn to the dataframe with NaN values.

```python
# dropping the Nan values
df = df.dropna(axis=1)
df.rename(columns = {'Experience' : 'Experience (in yrs)'}, inplace = True)
df
```

**Observation:** We drop the column with NaN values and rename the existing column name to the original name.

```python
# As we are converting a dictionary into DataFrame, we have to convert the data types of the columns from object (by default) into their respective data types
convert_dict = {'EMP ID': int,
                'Age': int,
                'Salary': int,
                'Experience (in yrs)': int
                }
 
df = df.astype(convert_dict)
```

```python
df.dtypes
```

**Observation:** As the dataframe was created from a dictionary, the data types of all the columns was set to 'objects' by default. So, we convert them to their respective types here.

```python
# User-Defined function for fiding the 10 smallest numbers in a column
min_empID=[]
def smallest():
    col = df2['EMP ID'].values.tolist()
    
    for h in range(10):
        min1 = col[0]
        for i in range(len(col)):
            if col[i] < min1:
                min1 = col[i]     
        min_empID.append(min1)
        col.remove(min1)
    return min_empID
```

```python
smallest()
```

```python
# User-Defined function for fiding the 10 largest numbers in a column
max_empID=[]
def largest():
    col = df2['EMP ID'].values.tolist()
    
    for h in range(10):
        min1 = col[0]
        for i in range(len(col)):
            if col[i] > min1:
                min1 = col[i]     
        max_empID.append(min1)
        col.remove(min1)
    return max_empID
```
