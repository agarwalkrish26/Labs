```python
  # Importing all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('D:/Z/Downloads/Faculty.csv')
```

```python
df
```

```python
# Dropping the 'Unnmaed: 0' column
df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
df
```

## Visualizations

```python
# --- 1 ---
label = df['Designation'].unique()
data = [len(df[df['Designation'] == 'Senior Prof']), len(df[df['Designation'] == 'IT']), len(df[df['Designation'] == 'Staff']), len(df[df['Designation'] == 'Junior Prof'])]

# plotting graph
plt.pie(data, labels = label, autopct='%1.0f%%')
plt.show()
```

```python
data
```

*Inference:* Senior Professors are less in number but overall the number of faculties belonging to each section are almost same in number.

```python
# --- 2 ---
sns.scatterplot(x=df['EMP ID'], y = df['Salary'], hue = df['Experience (in yrs)'])
```

*Inference:*

```python
# --- 3 ---
plt.hist(df['Experience (in yrs)'], width = 0.7)
```

*Inference:* Majority of the Faculties have atleast minimum five years of prior work experience.

```python
# --- 4 ---
sns.set(rc={'figure.figsize':(10, 10)})
sns.violinplot(x=df.Gender, y=df.Salary, data = df)
```

```python
var1 = df[df['Gender']=='Female']['Salary'].mean()
var2 = df[df['Gender']=='Male']['Salary'].mean()
var3 = df[df['Gender']=='Others']['Salary'].mean()

print("Female: ", var1, "\nMale: ", var2, "\nOthers: ", var3)
```

*Inference:* It is observed that Others have earned comparatively more in general than the other genders.

```python
# --- 5 ---
plt.hist(df['Age'], width = 0.7)
```

*Inference:* It is observed that majority of the Faculties are between the age of 50 to 55.

```python
# --- 6 ---
graph = sns.catplot(x = 'Age', y = 'Experience (in yrs)', hue = 'Gender', data = df)
```

```python
# --- 7 ---
data = df['Salary']
fig = plt.figure(figsize =(10, 7))

plt.boxplot(data)

plt.show()
```

```python
var4 = df['Salary'].min()
var5 = df['Salary'].max()
var6 = df['Salary'].median()

print("Min Salary: ", var4, "\nMax Salary: ", var5, "\nMedian Salary: ", var6)
```

*Inference:* The overall Salary of the Faculties range from approx. Rs.30,000 to approx. Rs.2,00,000 with a median of approx. Rs.1,15,000.
