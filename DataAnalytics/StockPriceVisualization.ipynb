```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  ------------------------------ Q2) Import the dataset onto python as a Pandas DataFrame as df ------------------------------
df = pd.read_csv('D:/Z/Downloads/Lab5/HCLTECH.csv')
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
#  ---------------------- Q3) Take any one year's data from the datasets, and name it as stockprice_one_year. Use this newly created DataFrame for the future analysis. ----------------------
stockprice_one_year = df[(df['Date'] > '2016-03-31') & (df['Date'] < '2017-04-01')]
stockprice_one_year
```

*Observation:* Double indexing the main dataframe and slicing out the row values which fall under the financial year 2016-17.

```python
# ---------------------- Q4) Compare stockprice_one_year['feature_name'].plot() vs df['feature_name'].plot() ----------------------

# Plotting the HIGH values of the original DataFrame
plt.figure(figsize = (15,7))
plt.title('High Values of HCL during FY 2000 to 2021')
plt.plot(df['Date'], df['High'])
plt.legend()
```

```python
# Plotting the HIGH values of 
plt.figure(figsize = (15,7))
plt.title('High Values of HCL during FY 2016-17')
plt.plot(stockprice_one_year['Date'], stockprice_one_year['High'])
plt.legend()
```

*Observation:* Firstly, plotting the High values of the all the row values spreaded out between the years 2000 to 2021. Secondly, plotting the High values of all the row values between the financial year 2016-17.

```python
# ---------------------- Q5) Check whether matplotlib's plotting is better in looks. Once the plotting is done, check what is the impact in Plots, if you execute the line of code: plt.style.use('fivethirtyeight') ----------------------
plt.figure(figsize = (15,7))
stockprice_one_year['Open'].plot()
plt.style.use('fivethirtyeight')
```

```python
stockprice_one_year['Date'][stockprice_one_year['Open'] == stockprice_one_year['Open'].max()]
```

```python
stockprice_one_year['Date'][stockprice_one_year['Open'] == stockprice_one_year['Open'].min()]
```

*Observation:* Plotting the High values of all the row values between the financial year 2016-17 and noticing that the highest value peaked on 2017-03-31 and the lowest value peaked on 2016-05-11 and 2016-05-12. 

```python
# ---------------------- Q6) Find the correlation between the features. Make a Correlation Heatmap. Write your observations about the same. ----------------------

plt.figure(figsize = (10, 5))
dataplot = sns.heatmap(stockprice_one_year.corr(), cmap="YlGnBu", annot=True)
plt.show()
```

*Observation:* Columns --> 'Prev Close', 'Open', 'High', 'Low', 'Last',
       'Close', 'VWAP' are highly correlated amongst each other with the correlation of almost 1. Also, 'Deliverable Volume'-'Volume' and'Derivable Volume'-'Turnover' are highly correlated. Rest of the data is weakly correlated.

```python
# ---------------------- Q7) How can you plot multiple features in a single plot? For instance, use a red line for "LOW", green line for "HIGH", and black line for "CLOSE". ----------------------

plt.figure(figsize = (10, 5)) 
stockprice_one_year['Low'].plot(label='Low', color = 'red')
stockprice_one_year['High'].plot(label='High', color = 'green')
stockprice_one_year['High'].plot(label='Close', color = 'black', linewidth = 10, alpha = 0.5)
plt.title('Comparing Multiple Features')
plt.legend()
```

```python
stockprice_one_year['Date'][stockprice_one_year['Low'] == stockprice_one_year['Low'].min()]
```

```python
stockprice_one_year['Date'][stockprice_one_year['High'] == stockprice_one_year['High'].max()]
```

*Observation:* Plotting the High and Low values of all the row values between the financial year 2016-17 and noticing that the highest value peaked on 2017-03-31 and the lowest on 2016-05-11.

```python
# ---------------------- Q8) What does plt.fill_between() do? can you shade the region between LOW and HIGH? ----------------------
plt.figure(figsize = (10, 5)) 
plt.plot(stockprice_one_year['Date'], stockprice_one_year['High'], color = 'magenta', linewidth = 1, label = 'High')
plt.fill_between(stockprice_one_year['Date'], stockprice_one_year['High'], stockprice_one_year['Low'])
plt.plot(stockprice_one_year['Date'], stockprice_one_year['Low'], color = 'black', linewidth = 1, label = 'Low')
```

Plotting the High and Low values of all the row values between the financial year 2016-17 and noticing that the highest value peaked on 2017-03-31 and the lowest on 2016-05-11. The graph also portrayed the deviation between both the line graphs.

```python
# ---------------------- Q9) Take the CLOSE price, can you give a colour for the first half of the year, and a different colour for second half ----------------------
df2 = stockprice_one_year[:124]
df3 = stockprice_one_year[124:]
```

```python
plt.figure(figsize = (10, 5))
df2['Close'].plot(label = 'First Half', color = 'Magenta', linewidth = 3)
df3['Close'].plot(label = 'Second Half', color = 'Black', linewidth = 3)
plt.title('Plotting 2 halves of the dataframe')
plt.legend()
```

*Observation:* Manually splitting the dataframe into 2 equal halves and then plotting the same onto the same graph. It makes it more readable and understandable.

```python
# ---------------------- Q10) Does BoxPlot signify something for the sampled dataset? ----------------------
df4 = stockprice_one_year.drop(['Date', 'Symbol', 'Series', 'Volume', 'Turnover', 'Trades', 'Deliverable Volume', '%Deliverble'], axis=1)
sns.boxplot(data = df4)
```

*Observation:* Box plot gives out the range of spreadness of the data. It also gives out Q1, Q2 (median), Q3, Q4, minimum value, maximum value and the outliers (if any). In thsi case, I had to omit out the columns with exponential values in order to plot the graph.
