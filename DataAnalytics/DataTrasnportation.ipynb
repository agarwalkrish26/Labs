### --- Importing necessary libraries ---

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import groupby
from operator import itemgetter
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from pandas_profiling import ProfileReport
```

### --- Importing Dataset onto Python ---

```python
df = pd.read_excel('D:/Z/Downloads/TreasureHuntDataset.xlsx')
```

```python
df
```

### --- Data Wrangling ---

#### Removing Rows with invalid Registration Numbers

```python
df.drop(2,axis=0,inplace=True)
```

#### Sorting the DataFrame on the basis of Registration Number

```python
# Converting the data type to integer
df['Register Number'] = df['Register Number'].astype(int)
```

```python
df = df.sort_values(by=['Register Number'], ascending=True)
```

```python
# As we have removed an instance from the data frame, we reset the index
df.reset_index(inplace = True, drop = True)
```

#### Removing of invalid and multiple entries of Registration Number

Note: 1. We are keeping the most recent entry on the basis of timestamp and removing other entries under the same registration numbers.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;2. We are also keeping entries with not NaN values in case of multiple entries to make better insights.

```python
duplicate = df[df.duplicated(['Register Number'])]
```

```python
duplicate
```

```python
# Removing the multiple entries from their index values
df.drop([0, 3, 26, 43, 55, 59, 60],axis=0,inplace=True)
```

#### Renaming columns

```python
df.rename(columns = {'We made _______ during the event':'Communication'}, inplace = True)
df.rename(columns = {'On a scale of 1 to 5, How much did you like the event?':'Event (/5)'}, inplace = True)
df.rename(columns = {'Did you find the Clues Difficult?':'Clues (/5)'}, inplace = True)
df.rename(columns = {'Did you find the First Round (Quiz) Easy?':'First Round (/5)'}, inplace = True)
df.rename(columns = {'Enter your Team Mates Register Numbers (Separated by Commas)':'Team Mates'}, inplace = True)
df.rename(columns = {'How many team mates did you know before?':'Team Mates Known Prior'}, inplace = True)
df.rename(columns = {'How many of you were in the Team?':'Total Members'}, inplace = True)
df.rename(columns = {'How many team mates spoke the same language as yours?':'Same Language'}, inplace = True)
df.rename(columns = {'Count of Team Mates from 1 BSc EA':'1BSc EA'}, inplace = True)
df.rename(columns = {'Count of Team Mates from 1 BSc DS':'1BSc DS'}, inplace = True)
df.rename(columns = {'Count of Team Mates from 3 BSc EA':'3BSc EA'}, inplace = True)
df.rename(columns = {'Count of Team Mates from 3 BSc DS':'3BSc DS'}, inplace = True)
df.rename(columns = {'Count of Team Mates from 5 BSc EA':'5BSc EA'}, inplace = True)
df.rename(columns = {'Count of Team Mates from 5 BSc DS':'5BSc DS'}, inplace = True)
df.rename(columns = {'Count of Team Mates from MSc':'MSc'}, inplace = True)
```

```python
df.reset_index(inplace = True, drop = True)
```

#### Replacing NaN values with 0  
Note: Since a lot of entries consist of NaN values, it could be a tedious task to fill in those values. Hence, we are filling those values simply with 0.

```python
df = df.fillna(0)
```

#### Cleaning the data column wise

Team Mates Known Prior

```python
df['Team Mates Known Prior'].unique()
```

```python
# instances that have to be converted into zero
team_prior_zero = ['NONE', 'None', 'Zero', 'none', 'volunteers', '-', '------Null']
```

```python
# instances that have to be converted into one
team_prior_one = ['One ', 'one']
```

```python
# the match method replaces the value wit the given value if the condition if satisfied
for i in team_prior_zero:
    df['Team Mates Known Prior'].mask((df['Team Mates Known Prior'] == i) ,0, inplace=True)
```

```python
for i in team_prior_one:
    df['Team Mates Known Prior'].mask((df['Team Mates Known Prior'] == i) ,0, inplace=True)
```

Same Language

```python
df['Same Language'].unique()
```

```python
# instances that have to be converted into zero
same_lang_zero = ['None', 'None (all English (', 'Everyone spoke in English ', 'volunteers', '-', '------Null']
```

```python
# instances that have to be converted into one
same_lang_one = ['One']
```

```python
# instances that have to be swapped with their total members 
same_lang_all = ['All', 'All have communicated in English sometimes hindi', 'Everyone', 'All of them', 'all of them', 'Everyone spoke the same language', 'all 5', 'all', 'Everytime']
```

```python
# instances that have to be swaapped with the average of the column 
same_lang_avg = ['few ']
```

```python
for i in same_lang_zero:
    df['Same Language'].mask((df['Same Language'] == i) ,0, inplace=True)
```

```python
for i in same_lang_one:
    df['Same Language'].mask((df['Same Language'] == i) ,1, inplace=True)
```

```python
for i in same_lang_all:
    df['Same Language'].mask((df['Same Language'] == i) , df['Total Members'], inplace=True)
```

```python
for i in same_lang_avg:
    df['Same Language'].mask((df['Same Language'] == i) , 3, inplace=True) # 3 being the average of ceil(0+1+...5/6)
```

#### Changing the data types of the columns

```python
df.dtypes
```

```python
# Changing the data type of columns to their respective data type
df['Team Mates Known Prior'] = df['Team Mates Known Prior'].astype(int)
df['Total Members'] = df['Total Members'].astype(int)
df['Same Language'] = df['Same Language'].astype(int)
df['1BSc EA'] = df['1BSc EA'].astype(int)
df['1BSc DS'] = df['1BSc DS'].astype(int)
df['3BSc EA'] = df['3BSc EA'].astype(int)
df['3BSc DS'] = df['3BSc DS'].astype(int)
df['5BSc EA'] = df['5BSc EA'].astype(int)
df['5BSc DS'] = df['5BSc DS'].astype(int)
df['MSc'] = df['MSc'].astype(int)
```

```python
df.dtypes
```

#### Dropping of unnecessary columns 

```python
df = df.drop(['Timestamp', 'Team Mates', 'Communication'], axis=1)
```

```python
df.head()
```

#### Label Encoding

```python
# Label Encoding the Gender Column
gender_le = LabelEncoder()
df['Gender'] = gender_le.fit_transform(df['Gender']) # Female --> 0, Male --> 1
```

```python
df.head()
```

#### One Hot Encoding

```python
# One Hot Encoding 'Final Position' Column
one_hot_encoder_finalpos = OneHotEncoder()
final_position_reshaped = np.array(df['Final Position']).reshape(-1, 1)
final_position_values = one_hot_encoder_finalpos.fit_transform(final_position_reshaped)
```

```python
final_pos_col = df['Final Position'].unique()
final_position_df = pd.DataFrame(final_position_values.toarray(), columns = final_pos_col)
```

```python
df_final = pd.concat([df, final_position_df], axis = 1)
df_final = df_final.drop(['Final Position'], axis = 1)
```

```python
df_final.head()
```

#### Scaling

```python
scaler_one = MinMaxScaler()
df_final[['Event (/5)', 'Clues (/5)', 'First Round (/5)']] = scaler_one.fit_transform(df_final[['Event (/5)', 'Clues (/5)', 'First Round (/5)']])
```

```python
df_final.head()
```

#### Normalization

```python
scaler_two = StandardScaler()
df_final[['1BSc EA', '1BSc DS', '3BSc EA', '3BSc DS', '5BSc EA', '5BSc DS']] = scaler_two.fit_transform(df_final[['1BSc EA', '1BSc DS', '3BSc EA', '3BSc DS', '5BSc EA', '5BSc DS']])
```
