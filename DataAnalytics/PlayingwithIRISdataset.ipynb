```python
import pandas as pd
from sklearn import datasets
import numpy as np
```

**Q1) Import the Toy IRIS Dataset**

```python
iris = datasets.load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                  columns= iris['feature_names'] + ['Species'])

# Observation: We infer from this that there are multiple ways to import the IRIS dataset but here I take the NumPy approach and add all the feature_names and target_names.
# I also understood the concepts of target variable and feature variables in real life ML problems. Along with that, basic information about the dataframe viz it has five columns 
# named ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'Species'] and 150 rows.
```

**Q2) Print the Head and Tail of the Dataset (first 5 and last 5 entries)**

```python
#df.head() and df.tail() gives the same output as below
df

# Observation: Different Errors that people run into. They might not have converted their dataset into a dataframe. 
```

**Q3) Create a New Column named "Average" and find the Average of existing four features and store the same**

```python
#Can also use df.mean(axis=1)
df['Average'] = (df['sepal length (cm)'] + df['sepal width (cm)'] + df['petal length (cm)'] + df['petal width (cm)'])/4 


# Observation: Using of simple operators
```

```python
df
```

**Q4) Create a New Column named "Sum" and find the Sum of existing four features and store the same**

```python
df['Sum'] = (df['sepal length (cm)'] + df['sepal width (cm)'] + df['petal length (cm)'] + df['petal width (cm)'])

# Observation: Using of simple operators
```

```python
df
```

**Q5) Check how many entires of "IRIS Setosa", "IRIS Virgnica" and "IRIS Versicolor" are present in the Dataset**

```python
# ['setosa' 'versicolor' 'virginica'] --> [0, 1, 2]

for i in range(len(df["Species"])):
    if df["Species"][i] == 0:
        df["Species"][i] = 'setosa'
    elif df["Species"][i] == 1:
        df["Species"][i] = 'versicolor'
    elif df["Species"][i] == 2:
        df["Species"][i] = 'virginica'
```

```python
df['Species'].value_counts()

# Observation: How menu-driven program was impliable into this situation and the mapping of values to their names wasn't a complex task. Overriding of variables.
```

**Q6) Find the Descriptive Statistics about all the Six Columns (existing four + two made-up features)**

```python
df.describe()

# Observation: The .describe() function provides the user a basic statistic summary of the dataframe viz count, mean, std, quartiles, etc. It also allowed me to come across multiple 
# in-built functions which the pandas library offers.
```

**Q7) Find the entries whose Sum Values are more than a User Input**

```python
print("-"*20)
option = float(input("Enter a number: "))
print("-"*20, "\n")

df[df['Sum']>option]

# Observation: There exists 27 entries which satisfies the given condition. Double Indexing.
```

**Q8) Find the entries whose Sepal_Length is greater than Petal_Width**

```python
df[df['sepal length (cm)']>df['petal width (cm)']]

# Observation: All the entries have their sepal lengths greater than petal width. Double Indexing.
```
