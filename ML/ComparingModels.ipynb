#### Q1) Make two classification datasets, having more than 50% informative features and a minimum of 2500 rows.

```python
# Importing necessary libraries/modules
from sklearn.datasets import make_classification
from sklearn.datasets import make_moons
import pandas as pd
import random
```

```python
random.seed(1)
X1, y1 = make_classification(n_samples=5000, n_features=10, n_informative=6, n_redundant=0, n_classes=2, class_sep=1.0, scale = 20, random_state=42)
```

```python
random.seed(2)
X2, y2 = make_classification(n_samples=5000, n_features=12, n_informative=7, n_redundant=0, n_classes=2, class_sep=1.0, scale = 45,  random_state=93)
```

```python
# Generating feature names for df1
feature_names1 = []
for i in range(len(X1[0])):
    feature_names1.append("F"+str(i+1))
    
# Generating feature names for df2
feature_names2 = []
for i in range(len(X2[0])):
    feature_names2.append("F"+str(i+1))
```

```python
df1 = pd.DataFrame(X1, columns = feature_names1)
df1['Target'] = y1
df1_ = df1.copy() # Creating a copy for further questions
df1
```

```python
df2 = pd.DataFrame(X2, columns = feature_names2)
df2['Target'] = y2
df2_ = df2.copy() # Creating a copy for further questions
df2
```

#### Q2) Do some kind of Scaling on the first 5 features.

```python
# Importing all the necessary libraries
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
# Apllying standard scaler on the first 3 features on df1
standard_scaler1 = StandardScaler()
df1[['F1', 'F2', 'F3']] = standard_scaler1.fit_transform(df1[['F1', 'F2', 'F3']])
```

```python
# Apllying MinMax scaler on the 4th and 5th on df1
minmax_scaler1 = StandardScaler()
df1[['F4', 'F5']] = standard_scaler1.fit_transform(df1[['F4', 'F5']])
```

```python
# Apllying MinMax scaler on the 4th and 5th on df2
standard_scaler2 = StandardScaler()
df2[['F4', 'F5']] = standard_scaler2.fit_transform(df2[['F4', 'F5']])
```

```python
# Apllying MinMax scaler on the first 3 features on df2
minmax_scaler2 = StandardScaler()
df2[['F1', 'F2', 'F3']] = standard_scaler2.fit_transform(df2[['F1', 'F2', 'F3']])
```

#### Q3) Compare the suitability of above datasets when performing Logistic Regression and Support Vector Machines.

##### Hyperparameter Tuning for scaled dataframes

```python
# Importing all the necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Defining the parameters
dfs = [df1, df2]
models = [SVC(), LogisticRegression()]
random_states = [325, 25, 23, 532, 98]
df_compare = pd.DataFrame(columns=['Dataframe', 'Model', 'Random State', 'Accuracy', 'F1 Score'])
```

```python
# Hyper parameter tuning for dataframes with scaling
for i in range(len(dfs)):
    for j in range(len(models)):
        for k in random_states:
            accuracy_dict = {}
            
            X_train, X_test, y_train, y_test = train_test_split(dfs[i].drop(['Target'], axis = 1), dfs[i]['Target'], test_size=0.3, random_state=42)
            model = models[j]
            model.fit(X_train, y_train)
            model.set_params(random_state = k)
            y_pred = model.predict(X_test)
            accuracy = accuracy_score(y_test, y_pred)
            f1_score_ = f1_score(y_test, y_pred)
            
            accuracy_dict['Dataframe'] = "df"+str(i+1)
            accuracy_dict['Model'] = str(models[j])
            accuracy_dict['Random State'] = k
            accuracy_dict['Accuracy'] = accuracy
            accuracy_dict['F1 Score'] = f1_score_
            
            df_compare = df_compare.append(accuracy_dict, ignore_index = True)
```

```python
# Removing random state from model name
for i in range(len(df_compare['Model'])):
    df_compare['Model'][i] = df_compare['Model'][i].split("(")[0]
```

```python
df_compare
```

##### Visualization

```python
# --- Models vs Accuracy ---
sns.catplot(data = df_compare, x = "Model", y = "Accuracy", kind = "box")
plt.show()
```

*Inference*: It can be observed that SVC has overall better prediction status and higher accuracy when compared to LogisticRegression.

```python
# --- DF vs Accuracy ---
sns.catplot(data = df_compare, x = "Dataframe", y = "Accuracy", kind = "box")
plt.show()
```

*Inference*: It is observed that df2 has higher accuracy scores than df1.

```python
# --- DF vs Accuracy vs Model ---
sns.catplot(data = df_compare, x = 'Dataframe', y = 'Accuracy', hue = "Model")
plt.show()
```

*Inference*: It is can be stated that both the classifiers are a good fit for the data but SVC has excellent in the second dataframe whereas LogisticRegression has performed well in both the cases. 

#### Q4) Check whether the applied transformation have an impact on the final quality of Classification Model, and present your findings neatly.

```python
# Defining the parameters
dfs = [df1_, df2_] # unscaled dataframes
models = [SVC(), LogisticRegression()]
random_states = [325, 25, 23, 532, 98]
```

```python
# Hyper parameter tuning for dataframes without scaling
for i in range(len(dfs)):
    for j in range(len(models)):
        for k in random_states:
            accuracy_dict = {}
            
            X_train, X_test, y_train, y_test = train_test_split(dfs[i].drop(['Target'], axis = 1), dfs[i]['Target'], test_size=0.3, random_state=42)
            model = models[j]
            model.fit(X_train, y_train)
            model.set_params(random_state = k)
            y_pred = model.predict(X_test)
            accuracy = accuracy_score(y_test, y_pred)
            f1_score_ = f1_score(y_test, y_pred)
            
            accuracy_dict['Dataframe'] = "df_"+str(i+1)
            accuracy_dict['Model'] = str(models[j])
            accuracy_dict['Random State'] = k
            accuracy_dict['Accuracy'] = accuracy
            accuracy_dict['F1 Score'] = f1_score_
            
            df_compare = df_compare.append(accuracy_dict, ignore_index = True)
```

```python
# Removing random state from model name
for i in range(len(df_compare['Model'])):
    df_compare['Model'][i] = df_compare['Model'][i].split("(")[0]
```

```python
df_compare
```

##### Visualization

```python
# --- DF vs Accuracy ---
sns.catplot(data = df_compare, x = "Dataframe", y = "Accuracy", kind = "box")
plt.show()
```

*Inference*: It is observed that the dataframes which were not scaled have better accuracy scores than the scaled dataframes. 

```python
# --- DF vs Accuracy vs Model ---
sns.catplot(data = df_compare, x = 'Dataframe', y = 'Accuracy', hue = "Model")
plt.show()
```

*Inference*: It is can be stated that both the classifiers are a good fit for the data but SVC has performed excellently in df_2 whereas LogisticRegression has performed well enough in all the cases.
